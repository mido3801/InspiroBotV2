```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
from wordcloud import WordCloud,STOPWORDS

import keras
from keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences
from keras import layers
from keras.models import Sequential
import keras.utils as ku
from keras.callbacks import EarlyStopping

import nltk
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('wordnet')
nltk.download('words')
```

DATA CLEANING:

Using Excel, I replace all characters of the form â€. This is an issue that some of the characters in the data set are in ISO-8859-1 but we are reading them using the UTF-8 encoding. Thus, we get some weird artifacts like that. Using this website, https://www.i18nqa.com/debug/utf8-debug.html, I changed all the ISO characters into their UTF equivalents. 

Additionally, there are some data points that are made up entirely of ISO characters and for simplicity's sake, I simply removed those from the data set.

Throughout this process, we notice that some quotes use non-english characters. We remove this later on through code.

```python
path = "C:/CU FALL 2020/CSCI 4622/final/archive/quotes_inspirational.csv"
quotes_frame = pd.read_csv(path)

quote_deduped = quotes_frame['Quote'].drop_duplicates()

all_quotes = list(quote_deduped)
print(len(all_quotes))
print(all_quotes[:3])
```

```python
words = set(nltk.corpus.words.words())

quotes_filtered = []
quotes_len = []
noneng = False

for s in all_quotes:
    noneng = False
    
    for w in nltk.wordpunct_tokenize(s):
        if not (w.lower() in words or not w.isalpha()):
            noneng = True
    if not noneng:
        quotes_filtered.append(s)
        quotes_len.append(len(nltk.wordpunct_tokenize(s)))

        
print(len(quotes_filtered))
print(quotes_filtered[:3])
```

```python
from matplotlib import pyplot as plt
from matplotlib.ticker import PercentFormatter

plt.hist(quotes_len, bins=20, alpha=0.5, histtype='bar', ec='black')
plt.title("Histogram of Quote Lengths with 20 bins")
plt.xlabel("Quote Lengths")
plt.ylabel("# of Quotes")
plt.savefig("quote_len_hist.jpg")

plt.show()
```

We can see that this filtering is very aggresive, as some of the sentences that are in english get removed if they are not in NLTK's english wordlist. The beneifit is that we know there are no quotes with nonenglish words.

```python
tokenizer = Tokenizer()

def generate_sequences(text):
    tokenizer.fit_on_texts(text)
    total_words = len(tokenizer.word_index)+1
    print("Total unique words in the text: " + str(total_words))
    
    input_sequences = []
    for line in text:
        seq = tokenizer.texts_to_sequences([line])[0]
        for i in range(1, len(seq)):
            ngram_seq = seq[:i+1]
            input_sequences.append(ngram_seq)
            
    return input_sequences, total_words

# Generating sequences
input_sequences, total_words = generate_sequences(quotes_filtered)
word_set = list(tokenizer.word_index.keys())

input_sequences[:5]
```

```python
print(tokenizer.word_index)
```

```python
def generate_input_sequence(input_sequences):
    maxlen = max([len(x) for x in input_sequences])
    input_sequences = pad_sequences(input_sequences, maxlen=maxlen)
    predictors, label = input_sequences[:, :-1], input_sequences[:, -1]
    label = ku.to_categorical(label, num_classes=total_words)
    return predictors, label, maxlen

predictors, label, maxlen = generate_input_sequence(input_sequences)

print(len(predictors))
print(len(predictors[0]))

print(len(label))
print(len(label[0]))
```

```python
# Building the model
embedding_dim = 64

def create_model(maxlen, embedding_dim, total_words):
    model = Sequential()
    model.add(layers.Embedding(total_words, embedding_dim, input_length = maxlen))
    model.add(layers.LSTM(128, dropout=0.2))
    model.add(layers.Dense(total_words, activation='softmax'))
    
    # compiling the model
    model.compile(loss='categorical_crossentropy', optimizer='adam')
    return model

model = create_model(maxlen, embedding_dim, total_words)
model.summary()
```

```python
model.fit(predictors, label, epochs=50, batch_size=64)

model.save("Quotes_generator.h5")
```

```python
from keras.models import load_model

path = "C:/CU FALL 2020/CSCI 4622/final/Quotes_generator.h5"

Quotes_gen = load_model(path)
```

```python
def generate_quote(seed_text, num_words, model, maxlen):
    for i in range(num_words):
        tokens = tokenizer.texts_to_sequences([seed_text])[0]
        tokens = pad_sequences([tokens], maxlen=maxlen, padding='pre')
        
        predicted = model.predict_classes(tokens)
        
        output_word = ''
        
        for word, index in tokenizer.word_index.items():
            if index == predicted:
                output_word = word
                break
        seed_text = seed_text + " " + output_word
    
    return seed_text
```

```python
def generate_random_quote(model, maxlen):
    seed_index = np.random.randint(0, total_words)
    seed_text = word_set[seed_index]
    num_words = np.random.randint(5, 30)
    
    for i in range(num_words):
        tokens = tokenizer.texts_to_sequences([seed_text])[0]
        tokens = pad_sequences([tokens], maxlen=maxlen, padding='pre')
        
        predicted = model.predict_classes(tokens)
        
        output_word = ''
        
        for word, index in tokenizer.word_index.items():
            if index == predicted:
                output_word = word
                break
        seed_text = seed_text + " " + output_word
        
    return seed_text
```
